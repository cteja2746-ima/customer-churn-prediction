```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data/WA_Fn-UseC_-Telco-Customer-Churn.csv")

df.head()
```

```python
print(df.shape)

print(df.columns)

df.info()
```

```python
df.isnull().sum()
```

```python
df['TotalCharges'] = pd.to_numeric(
    df['TotalCharges'],
    errors='coerce'
)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
sns.countplot(x='Churn', data=df)

plt.title('Customer Churn Distribution')

plt.show()
```

```python
sns.countplot(
    x='Contract',
    hue='Churn',
    data=df
)

plt.xticks(rotation=45)

plt.title('Contract Type vs Churn')

plt.show()
```

```python
sns.boxplot(
    x='Churn',
    y='tenure',
    data=df
)

plt.title('Tenure vs Churn')

plt.show()
```

```python
sns.boxplot(
    x='Churn',
    y='MonthlyCharges',
    data=df
)

plt.title('Monthly Charges vs Churn')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
print(df.head())
```

```python
churn_rate = df['Churn'].value_counts(normalize=True) * 100

print(churn_rate)
```

```python
sns.countplot(
    x='InternetService',
    hue='Churn',
    data=df
)

plt.title('Internet Service vs Churn')

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    x='PaymentMethod',
    hue='Churn',
    data=df
)

plt.xticks(rotation=45)

plt.title('Payment Method vs Churn')

plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(8,5))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```
